## Target encoding

```python
import category_encoders as ce
```

Przejdzmy do zakodowania zmiennej kategorycznej it_location przy pomocy target encoding

```python
target_encoder = ce.TargetEncoder(cols=['it_location'])
df_target_enc = target_encoder.fit_transform(df, df['price'])
df_target_enc['it_location']
```

Działanie dla targetowanej zmiennej ciągłej (może się trochę różnić zależnie od implementacji): w obrębie danej kategorii wyliczana jest średnia wartość dla cechy docelowej. Dzięki  temu uzyskujemy zamianę zmiennych kategorycznych na liczbowe, a co więcej, wartości te niosą dodatkową informację na temat przewidywanej cechy.


Plusy: 
- niesienie dodatkowej informacji
- pozybcie się zmiennej kateogrycznej
- brak dodania kolejnych wymiarów do naszych danych w wyniku kodowania

Minusy:
- zależne od dystrybucji targetu i niekiedy może byc przez to podatne na overfitting


### One-hot encoding


```python
onehot_encoder = ce.OneHotEncoder(cols=['main_category'])
onehot_df = onehot_encoder.fit_transform(df, df['price'])
onehot_df
```

W wyniku kodowania następuje dodanie cech do tabeli dla każdej kategorii z osobna. Plusami może być zakres wartości zmiennych, które uzyskujemy, a minusem może być występujące niekiedy znaczne zwiększenie wymiarowości tabeli.

### Count Encoder

```python
count_encoder = ce.CountEncoder(cols=['main_category'])
count_df = count_encoder.fit_transform(df, df['price'])
count_df
```

Działanie count encodera opiera sie na wykorzystaniu liczby wystąpień danej cechy w obrębie zbioru. Trzeba jednak mocno  uważać na ten sposób kodowania i sens jego użycia. Nie zawsze może on być dobry oraz czasem może być warto przeskalować otrzymane wyniki do pewnego zakresu. (mając też na względzie precyzję obliczeń jeśli jedna z klas jest znacznie większa od innej itp.)

### LeaveOneOut

```python
encoder_LOO = ce.LeaveOneOutEncoder(cols=["main_category"])
LOO_df = encoder_LOO.fit_transform(df, df['price'])
LOO_df
```

Zasada działania jest podobna do opisanego wyżej TargetEncodera, jednak różnicą jest pomijanie w obliczeniach wartości target, kodowanej w danym momencie próbki, w celu obniżenia wpływu działania outlierów.

Może teraz barplot z ilością kolumn, jednak wyniki są mocno przewidywalne i raczej wizualizacja wskaże cokolwiek jedynie na temat użycia onehot w tym konkretnym przypadku.

```python
sns.barplot(x=['onehot', 'count', 'LOO'], y=[len(x.columns) for x in [onehot_df, count_df, LOO_df]]).set_title("Wielkość dataframe po kodowaniu")
plt.show()
```

## Uzupełnianie braków

```python
from sklearn.impute import KNNImputer
from sklearn.metrics import mean_squared_error
import math
np.random.seed(140)

# przygotowanie ramki
df_small = df[['it_seller_rating', 'it_quantity', 'price']].sample(10000).reset_index(drop=True)
df_small
```

```python
def remove_random(df, column):
    indexes = np.random.choice(df.index.values, size=math.floor(len(df)/10))
    output_df = df.copy(deep=True)
    output_df.loc[indexes, column] = np.nan
    return output_df, indexes

def impute(df, n):
    imputer = KNNImputer(n_neighbors=n, weights="uniform")
    output_df = pd.DataFrame(imputer.fit_transform(df), columns=df.columns) 
    return output_df

def evaluate(basic, filled, indexes, column):
    return np.sqrt(mean_squared_error(basic.loc[indexes, column], filled.loc[indexes, column]))
    
```

##### Braki w jednej kolumnie

```python
wyniki=[]
for i in range(10):
    df_removed, indexes = remove_random(df_small, ['it_seller_rating'])
    df_filled = impute(df_removed,  10)
    wyniki.append(evaluate(df_small, df_filled, indexes, ['it_seller_rating']))
    
wyniki = np.array(wyniki)
```

```python
wyniki.std()
```

##### Braki w dwóch kolumnach

```python
wyniki_2=[]
for i in range(10):
    df_removed_seller, indexes = remove_random(df_small, ['it_seller_rating'])
    df_removed_quantity = remove_random(df_removed_seller,["it_quantity"])[0]#usuwamy tez losowo z drugiej kolumny
    df_filled = impute(df_removed_quantity,  10)
    wyniki_2.append(evaluate(df_small, df_filled, indexes, ['it_seller_rating']))

wyniki_2 = np.array(wyniki_2)   
print(wyniki_2)
```

```python
wyniki_2.std()
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots(figsize=(11,6))
bp = ax.boxplot([wyniki, wyniki_2], labels=['braki w rating','braki rating + quantity'])
plt.show()
```

## Podsumowanie

Jak można było przewidywać, przy brakach w jednej kolumnie algorytm poradził sobie nieco lepiej. Średnio metryka przy manipulacji ziarnem losowości była niższa (dla brakow w jednej kolumnie), jednak różnica ta najczęściej była niewielka. W drugim przypadku zmniejszała się wariancja. Widoczne jest to również po boxplotach i odległości między kwartylami.

Jeszcze z ciekawości, dla braków jedynie w jednej kolumnie wykonam boxploty, obrazujące wielkość metryki RSME w zależności od uwzględnianej ilości sąsiadów (do dokonania imputacji).

```python
wyniki_all=[]
for j in range (1,11):
    wyniki=[]
    for i in range(10):
        df_removed, indexes = remove_random(df_small, ['it_seller_rating'])
        df_filled = impute(df_removed,  j)
        wyniki.append(evaluate(df_small, df_filled, indexes, ['it_seller_rating']))
    wyniki_all.append(np.array(wyniki))
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots(figsize=(11,6))
bp = ax.boxplot(wyniki_all)
plt.ylabel("RSME")
plt.xlabel("Number of n-neighbors")
plt.show()
```

Jak widzimy liczba sąsiadów ma istotny wpływ na poprawność impotacji dokonanej przez algorytm. Poprawa jest bardziej dynamiczna szczególnie dla małych wartości parametru.
